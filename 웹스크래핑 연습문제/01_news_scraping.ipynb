```python
import requests
from bs4 import BeautifulSoup

req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
}

section_dict = {
    '정치': 'politics',
    '경제': 'economy',
    '사회': 'society',
    '세계': 'world',
    '문화': 'culture',
    '생활': 'life',
    'IT/과학': 'tech',
    '인물': 'people'
}

def print_news(section_name):
    if section_name not in section_dict:
        print(f"[ERROR] '{section_name}' 는 유효한 섹션명이 아닙니다.")
        return

    section_eng = section_dict[section_name]
    url = f'https://news.daum.net/{section_eng}'
    print(f"\n==== {section_name} 뉴스 ====")
    print(f"URL: {url}\n")

    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'

    if res.ok:
        html = res.text
        soup = BeautifulSoup(html, 'html.parser')

        li_tags = soup.select('ul.list_newsheadline2 li')

        if not li_tags:
            print("기사를 찾을 수 없습니다.")
            return

        for idx, li_tag in enumerate(li_tags, 1):
            a_tag = li_tag.find('a')
            link = a_tag['href'] if a_tag else "링크 없음"

            strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
            title = strong_tag.text.strip() if strong_tag else "제목 없음"

            print(f"{idx}. {title} {link}")
    else:
        print("페이지 요청 실패")

# 테스트
print_news('경제')
print_news('사회')
```
