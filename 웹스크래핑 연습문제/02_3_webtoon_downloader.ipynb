```python
import requests
from bs4 import BeautifulSoup
import os
import time

def download_one_episode(title, no, url):
    headers = {
        'referer': url,
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
    }
    res = requests.get(url, headers=headers)
    if res.status_code != 200:
        print(f"[{no}화] 요청 실패 (status: {res.status_code})")
        return

    soup = BeautifulSoup(res.text, 'html.parser')
    img_tags = soup.select("img[src*='IMAG']")

    if not img_tags:
        print(f"[{no}화] 이미지 없음")
        return

    save_dir = os.path.join("img", title)
    os.makedirs(save_dir, exist_ok=True)

    first_img_url = img_tags[0]['src']
    save_path = os.path.join(save_dir, f"{no:03d}.jpg") 

    img_res = requests.get(first_img_url, headers=headers)
    if img_res.ok:
        with open(save_path, 'wb') as f:
            f.write(img_res.content)
        print(f"[{no}화] 첫 이미지 저장 완료: {save_path}")

    print(f"[DONE] {no}화 저장 완료")

def download_range(title, title_id):
    Top_url = f"https://comic.naver.com/api/article/list?titleId={title_id}&page=1"
    headers = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
    }
    res = requests.get(Top_url, headers=headers)
    if res.status_code == 200:
        data = res.json()
        latest = data['articleList'][0]
        latest_no = latest['no']
        print(f"[Info] 최신화: {latest_no}")
    else:
        print("[Error] 최신화 정보를 가져오지 못했습니다.")
        return

    for no in range(1, latest_no):
        url = f"https://comic.naver.com/webtoon/detail?titleId={title_id}&no={no}"
        download_one_episode(title, no, url)
        time.sleep(1)

# 실행
download_range("배달왕", 823933)
```
